Q1: Write a python program to scrape data for ‚ÄúData Analyst‚Äù Job position in ‚ÄúBangalore‚Äù location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter ‚ÄúData Analyst‚Äù in ‚ÄúSkill, Designations, Companies‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the
location‚Äù field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
Note: All of the above steps have to be done in code. No step is to be done manually.

```python
import selenium                                 
from selenium import webdriver                  
import pandas as pd                              #create DataFrame
from selenium.webdriver.common.by import By     
import warnings                                  
warnings.filterwarnings("ignore")
import time                                      

#search for naukri.com in automated Chrome browser.
driver = webdriver.Chrome()
driver.get("https://www.naukri.com/") 
```

```python
#Enter Designation
designation=driver.find_element(By.CLASS_NAME,"suggestor-input")
designation.send_keys("Data Analyst")
```

```python
#Entering the location 
location=driver.find_element(By.XPATH, "/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location.send_keys("Bangalore")
```

```python
#Click Search 
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
#empty lists for title, location, company name and exp.
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#Scraping the job title.
title_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
```

```python
#Scraping the location.
location_tags=driver.find_elements(By.XPATH,"//span[@class='ellipsis fleft locWdth']")
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)

#Scraping the company names.  
company_tags=driver.find_elements(By.XPATH,"//a[@class='subTitle ellipsis fleft']")
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)

#Scraping the experience required.
experience_tags=driver.find_elements(By.XPATH,"//span[@class='ellipsis fleft expwdth']")
for i in experience_tags[0:10]:
    exp=i.text
    experience_required.append(exp)
```

```python
#Checking if the length of all the lists is same to create a data frame.
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
df= pd.DataFrame({'Title':job_title,'Location':job_location,'Company':company_name,'Experience':experience_required})
df
```

As shown in the above page you have to scrape the tick marked attributes. These are:
1. Rating
2. Review summary
3. Full review
4. You have to scrape this data for first 100reviews.

```python
import selenium                                  
from selenium import webdriver                   
import pandas as pd                             
from selenium.webdriver.common.by import By     
import warnings                                  
warnings.filterwarnings("ignore")
import time                                      

#oepn specificed URL in  automated chrome browser
driver = webdriver.Chrome()
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/%20itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market%20place=FLIPKART")
```

```python
# Creating empty lists for rating, summary on reviews
ratings = [] 
summary = []
reviews = []
```

```python
#To check the length of the 'rating' list
while len(ratings) < 100:
    rating = driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    for i in rating:
        r=i.text
        ratings.append(r)
        if len(ratings)>=100:
            break 
        
    try:
        button = driver.find_element(By.CLASS_NAME,"_1LKTO3")
        button.click()
    except NoSuchElementException:
        break    
    
```

```python
while len(summary)<100: #To check the length of the 'summary' list
    summary_ = driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    for i in summary_:
        s =i.text
        summary.append(s)
        if len(summary_)>=100:
            break 
    try:
        button = driver.find_element(By.CLASS_NAME,"_1LKTO3")
        button.click()
    except NoSuchElementException:
        break 
    
```

```python
# Looping until at least 100 reviews are collected
while len(reviews) < 100:
    # Finding all elements with the specified XPath for reviews and storing them in the 'review' list
    review = driver.find_elements(By.XPATH, '//div[@class="t-ZTKy"]')
    for i in review:
        r = i.text
        reviews.append(r)  # Adding the text of each review element to the 'reviews' list
        if len(reviews) >= 100:
            break  # If at least 100 reviews are collected, exit the loop

    try:
        # Finding the element with class name "_1LKTO3" (likely a "Load More" button) and clicking it to load more reviews
        button = driver.find_element(By.CLASS_NAME, "_1LKTO3")
        button.click()
    except NoSuchElementException:
        # If the "Load More" button is not found (indicating no more reviews to load), exit the loop
        break
```

```python
print(len(ratings),len(summary),len(reviews))
```

```python
#Creating a DataFrame using Pandas and calling it
df=pd.DataFrame({'Ratings':ratings,'Summary':summary,'Full Review':reviews})
df 
```

Q9: Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead,
Term of office, Remarks) from https://www.jagranjosh.com/.
This task will be done in following steps:
1. First get the webpagehttps://www.jagranjosh.com/
2. Then You have to click on the GK option
3. Then click on the List of all Prime Ministers of India
4. Then scrap the mentioned data and make theDataFrame.

```python
import selenium                                  
from selenium import webdriver                 
import pandas as pd                              
from selenium.webdriver.common.by import By      
import warnings                                  
warnings.filterwarnings("ignore")
#use to stop search engine for few seconds
import time     
#to create a DataFrame
import numpy as np                              
#Opening 'jagranjosh.com in automated chrome browser'
driver = webdriver.Chrome()
driver.get("https://www.jagranjosh.com/") 
```

```python
#Clicking on the hamburger on the top left corner of the page.
hamburger = driver.find_element(By.CLASS_NAME,"mob-icon")
hamburger.click()
```

```python
#Slecting 'GK' from the list
GK = driver.find_element(By.XPATH,'//a[@href="/general-knowledge?ref=nav_dd"]')
GK.click()
```

```python
# Selecting the tab 'IndiaGK'
select_tab = driver.find_element(By.XPATH,'//a[@href="#indiaGK"]')
select_tab.click()
```

```python
#Selecting 'List of Prime_ministers'
name_tag = driver.find_element(By.XPATH,'//a[@href="/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1"]')
name_tag.click()
```

```python
#creating empty list
pm_data = [] 
pm_table = driver.find_element(By.CLASS_NAME, "table-box")
rows = pm_table.find_elements(By.TAG_NAME, "tr")[1:] 
```

```python
#scapping the data 
for row in rows:
    cols = row.find_elements(By.TAG_NAME, "td")
    name = cols[1].text.strip() 
    born_dead = cols[2].text.strip() 
    term_of_office = cols[3].text.strip() 
    remarks = cols[4].text.strip()
    pm_data.append((name, born_dead, term_of_office, remarks)) 
```

```python
len(pm_data)
```

```python
driver.quit() #Close the automated chrome browser
```

```python
#Using the NumPy array as the data is stored in array format
data =np.array(pm_data)  
columns = ["Name", "Born-Dead", "Term of Office","Remarks"]
df = pd.DataFrame(pm_data, columns=columns)
df
```

Q10: Write a python program to display list of 50 Most expensive cars in the world (i.e.
Car name and Price) from https://www.motor1.com/
This task will be done in following steps:
1. First get the webpage https://www.motor1.com/
2. Then You have to type in the search bar ‚Äô50 most expensive cars‚Äô
3. Then click on 50 most expensive cars in the world..
4. Then scrap the mentioned data and make the dataframe.

```python
import selenium                                  
from selenium import webdriver                   
import pandas as pd                             
from selenium.webdriver.common.by import By      
import warnings                                  
warnings.filterwarnings("ignore")
#use to stop search engine for few seconds
import time                                      
#Opening motor1 in automated chrome browser
driver = webdriver.Chrome()
driver.get("https://www.motor1.com/") 
```

```python
#Entering the query in the search bar.
product = driver.find_element(By.XPATH,'/html/body/div[10]/div[2]/div/div/div[3]/div/div/div/form/input')
product.send_keys("50 most expensive cars") 
```

```python
#Clicking onthe search button
search = driver.find_element(By.XPATH,'//button[@class="m1-search-panel-button m1-search-form-button-animate icon-search-svg"]')
search.click() 
```

```python
#Clicking on the article
article = driver.find_element(By.XPATH,'/html/body/div[10]/div[9]/div/div[1]/div/div/div[2]/div/div[1]')
article.click() 
```

```python
#Creating empty lists
cars = [] 
prices = []
```

```python
#Scrapping the car names data.
car=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in car[0:50]:
    c=i.text
    cars.append(c) 
```

```python
#Scrapping  the price details.
price= driver.find_elements(By.XPATH, '/html/body/div[10]/div[7]/div[2]/div[1]/div[2]/div[1]/p/strong')

for i in price[0:50]:
    p = i.text
    prices.append(p)
```

```python
#Checking the lenght of both the lists
print(len(cars),len(prices))
```

```python

df=pd.DataFrame({'Cars': cars,'price':prices})
df
```
